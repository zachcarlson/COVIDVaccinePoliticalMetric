It is important to understand what these different numbers mean:

- `total_vaccinations`:  total number of **doses** given.  This number will not be equal to total number of people vaccinated or fully vaccinated as there are different vaccinate protocols depending on its manufacturer (e.g. 1 doses or 2).  
- `total_vaccinations_per_hundred`: `total-vaccinations` normalized to population.
- `daily_vaccinations_raw`: daily change in total number of doses given.
- `daily_vacccinations`: new doses given per day smoothed over 7 days.
- `daily_vaccinations_per_million`: `daily-vaccinations` normalized to population.
- `people_vaccinated`: total number of people given at least one doses.  If someone receives a second dose, this metric doesn't change.
- `people_vaccinated_per_hundred`: `people_vaccinated` normalized to population.
- `people_fully_vaccinated`: total number of people given all doses for their respective protocol.  If a person receives the first of 2 doses, this metric doesn't change.  If a person receives the second of 2 doses, this metric increases by 1.
- `people_fully_vaccinated_per_hundred`: `people_fully_vaccinated` normalized to population.
- `total_distributed`: cumulative counts of vaccine doses **shipped** as recorded by the CDC's Vaccine Tracking System.
- `total_distributed_per_hundred`: `total_distributed` normalized to population.
- `share_doses_used`: Fraction of doses given among those that were recorded as **shipped** by the CDC's Vaccine Tracking System.

Let's make a function `get_state_table()` that takes a df (like `vdata_final`) and a `location` value and returns a df of that `location` with a `DateTimeIndex`.

```python
from dateutil.parser import parse

def get_state_table(df, location):
    df_sub = df[df["location"] == location]
    df_sub = df_sub.set_index('date')
    df_sub.index = pd.Series(df_sub.index, dtype = "datetime64[ns]")
    return df_sub
```

Similar to Assignment 2 Module A, we'll create a function `plot_series()` that will plot the time series data of a given `metric` (e.g. `people_vaccinated`) from a dataframe `vdata` that has been passed through `get_state_table()`.  You can set a `start` and `stop` date, list the `title` and say whether you want a `legend` or not.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

def plot_series(vdata, states, metric, start, stop, title, legend = True):
        
    fig = plt.figure(figsize = (18,6))
    
    for state in states:
        df = get_state_table(vdata, state)
        _ = plt.plot(df.index, df[metric], label = df["location"][0], alpha = 0.5)
        
    plt.xlim([parse(start, yearfirst = True), parse(stop, yearfirst = True)])
    plt.ylabel(title, fontsize = 15)
    plt.xlabel("Date", fontsize = 15)
    if legend:
        plt.legend(fontsize = 15)
  
    return fig
    
```

```python
states_plot = ['Alaska', 'Texas', 'New York State', 'Michigan', 'Pennsylvania', 'Florida']
fig = plot_series(vdata = vdata_final, 
                  states = states_plot, 
                  metric = 'people_vaccinated_per_hundred', 
                  start = '2021-01-15', 
                  stop = '2021-06-01', 
                  title = "People Vaccinated Per Hundred")
```

There seems to be a big difference between people vaccinated per hundred on the latest date, `2021-05-09`.  Let's do some distribution analysis at this time point by making a function `standardize()` which standardizes a given `vdata` and returns it.

```python
import numpy as np

def standardize(vdata):
    vdata_standardize = (vdata - np.mean(vdata)) / np.std(vdata)
    return vdata_standardize
```

There appears to be two disributions, one > 45% vaccinated and one < 45% vaccinated.  Let's now import election data and see if we can see any differences.

The election data covers the U.S. Presidental election from 1976-2020.  It is provided by the M.I.T. Election Data and Science Lab and is provided [here](https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/42MVDX).  We'll save the file as `election_results.csv` and store it in `./data`.

First we'll filter out only data on `DEMOCRAT` and `REPUBLICAN`.

```python
election_results_raw = pd.read_csv("./data/election_results.csv")
political_parties = ["DEMOCRAT", "REPUBLICAN", "DEMOCRATIC-FARMER-LABOR"] #Minnesota had John Kerry listed as "DEMOCRATIC-FARMER-LABOR" in 2000.
election_results = election_results_raw[election_results_raw["party_detailed"].isin(political_parties)]
election_results = election_results[election_results["candidate"] != "OTHER"] #filter out write ins
election_results = election_results[election_results["writein"] != True] #filter out additional write ins
election_results.index = list(range(election_results.shape[0]))
```

Now we'll go through each election year and each state and assign a `1` for `REPUBLICAN` and `-1` for `DEMOCRAT` based off which had higher counts.

```python
election_years = list(election_results.year.unique())
for i, year in enumerate(election_years):
    election_years[i] = str(election_years[i])
election_states = list(election_results.state.unique())

#create a database to start election results
election_winners = pd.DataFrame(index = election_years, columns = election_states)

print(election_winners.iloc[:, 0:5])
print(election_winners.index)
```

```python
#write election_results file to debugging
election_results.to_csv(r'~/Google Drive/Grad School/DSCI 521/Project/data/election_results_FINAL.csv', index = False, header=True)
```

```python
democrat_names = ["DEMOCRAT", "DEMOCRATIC-FARMER-LABOR"]
for i in range(election_results.shape[0]): 
    #go through each row, in two-party pairs
    if ((i % 2 == 0) & (i < election_results.shape[0] -1)): 
        #checking if row is even, using this to analyze every other row
        i_year = str(election_results["year"][i]) #get year 
        j_state = str(election_results["state"][i]) #get state
        
        if ((election_results["candidatevotes"][i] > election_results["candidatevotes"][i+1]) 
            & (election_results["writein"][i] != True)):
            #even row is greater than odd row
            if election_results["party_detailed"][i] in democrat_names: 
                #party is democratic. assign a -1
                election_winners[j_state].loc[[i_year]] = -1
            else: #party is republican. assign a 1
                election_winners[j_state].loc[[i_year]] = 1
        else:
            #odd row is greater than even row
            if election_results["party_detailed"][i+1] in democrat_names:
                #party is democrat, assign -1
                election_winners[j_state].loc[[i_year]] = -1 
                
            else:
                #party is republican.  assign a 1
                election_winners[j_state].loc[[i_year]] = 1
            
```

This code correctly assigns a `1` or a `-1`, except for one case.  In the 2020 election, the District of Columbia data is written as a `writein`, which we have selectively removed.  Because the focus of this class is not on pre-processing, I will be hardcoding the value of a `-1`, for democrat.

```python
election_winners['DISTRICT OF COLUMBIA'].loc['2020'] = -1
```

Now that we have assigned a `1` or a `-1` for each state for each election year, let's get the sum so we get a metric of redness versus blueness for each state and then sort by this metric.

```python
election_winners_trans = np.transpose(election_winners)
election_winners_sum = np.sum(election_winners_trans, axis = 1)
election_winners_sum = election_winners_sum.sort_values(ascending = False)
print(election_winners_sum.head())
print(election_winners_sum.tail())
print(election_winners_trans.head())
```

It is well known California is a blue state, but according to these calculations it's a perfect neutral with a value of `0.0`.  This is because it was historically red and has voted democratic in the most recent elections.  

We need to add `weights` to each election year.  These `weights` should add more weight to more recent elections compared to older ones.  Let's do this calculation now.  There are 12 election years and we'll add weights ranging from `0.01` to `1`.

```python
weights = [0.001, 0.083, 0.17, 0.25, 0.33, 0.41, 0.583, 0.75, 0.83, 0.87, 0.92, 1]
```

```python
political_metric = []
state_names = []
for i in range(election_winners_trans.shape[0]):
    state = election_winners_trans.iloc[i,:].name
    metric = np.dot(election_winners_trans.iloc[i,:], weights)
    political_metric.append(metric)
    state_names.append(state)
    
data_tuples = list(zip(state_names, political_metric))
state_metrics = pd.DataFrame(data_tuples, columns=['State', 'Metric'])
```
